```python
with Session() as session:
    stmt = (
        select(Player)
        .where(
            and_(
                Player.active == True,
                or_(
                    Player.nickname.like("%a%"),
                    Player.nickname.like("%e%"),
                )
            )
            # and( x or( z / a ))
        )
        .order_by(Player.nickname)
    )

    players = session.scalars(stmt).all()


    for result in players:
        print(f"Player: {result.nickname}")

    print(result)
```
