```python
class player():
    def __init__(self, id, nickname, email, birth_date, registration_date, active, id_region):
        self.id = id, 
        self.nickname = nickname, 
        self.email = email, 
        self.birth_date = birth_date,
        self.registration_date = registration_date,
        self.active = active, 
        self.id_region = id_region



class player_profile():
    def __init__(self,id_playerprofile,  total_matches, total_wins, total_losses, total_points, ranking_position, last_update, id_player):
        self.id_playerprofile= id_playerprofile,
        self.total_matches= total_matches, 
        self.total_wins= total_wins, 
        self.total_losses= total_losses, 
        self.total_points= total_points, 
        self.ranking_position= ranking_position, 
        self.last_update= last_update,
        self.id_player= id_player



# Relacion n:m con atributos
class player_team():
    def __init__(self, id_team, id_player, join_date):
        self.id_team = id_team,
        self.id_player = id_player, 
        self.join_date = join_date



class region():
    def __init__(self,id_region, name, country_code ):
        self.id_region = id_region,
        self.name = name,
        self.country_code = country_code



class team():
    def __init__(self,id_team, name, creation_date, logo_url, active, id_region):
        self.id_team = id_team ,
        self.name = name ,
        self.creation_date = creation_date ,
        self.logo_url = logo_url ,
        self.active = active ,
        self.id_region = id_region 
        


class tournament():
    def __init__(self, name, start_date, end_date, prize_pool, max_teams, id_region, id_videogames_type, id_tournament_mode, id_tournament_status ):
        self.name = name,
        self.start_date = start_date,
        self.end_date = end_date,
        self.prize_pool = prize_pool,
        self.max_teams = max_teams,
        self.id_region = id_region,
        self.id_videogames_type = id_videogames_type,
        self.id_tournament_mode = id_tournament_mode,
        self.id_tournament_status = id_tournament_status

    

class match():
    def __init__(self, id_match, match_date, score_team_one, score_team_two, round_number, duration_minutes, id_team_1, id_team_2, id_team_winner, id_tournament):
        self.id_match= id_match,
        self.match_date= match_date,
        self.score_team_one= score_team_one,
        self.score_team_two= score_team_two,
        self.round_number= round_number,
        self.duration_minutes= duration_minutes,
        self.id_team_1= id_team_1,
        self.id_team_2= id_team_2,
        self.id_team_winner= id_team_winner,
        self.id_tournament= id_tournament



class video_game_type():
    def __init__(self,id_type, name, description):
        self.id_type =id_type ,
        self.name =name ,
        self.description =description 



class tournament_mode():
    def __init__(self, id_tourn_mode, name, description):
        self.id_tourn_mode = id_tourn_mode, 
        self.name = name, 
        self.description = description



class tournament_status():
    def __init__(self, id_tournament_status, name, description):
        self.id_tournament_status = id_tournament_status,
        self.name = name,
        self.description = description



```
